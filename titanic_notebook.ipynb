```python
df = pd.read_csv('/kaggle/input/titanic/train.csv')
Y = df.pop('Survived')
df.drop(['PassengerId', 'Name', 'Ticket'],axis=1, inplace=True)
print(Y)
df
```

```python
#vectorize Sex and Embarked Location
df['Sex'] = (df['Sex'] == 'male').astype(int)
mapping = {'S': 0, 'C': 1, 'Q': 2}
df['Embarked'] = df['Embarked'].map(mapping)
df
```

```python
#Now its time to look at "Cabin"
#Lots of NaN marks in the column, not really usable in its current state
#Heres what I wanna do instead, 
# 1. Split numbers from layers
# 2. Create a subtable where Cabin != NaN
# 3. Train a model to predict the cabin positions for all the other members
# 4. Fill in the NaN marks with the model predictions
# 5. Lastly use the resulting data set to train the final model

#but first lets take a look at the correlations, to do so, we will create a new df that doesnt have the cabins
# since we cant draw a correlation matrix from a string 

newdf = df.drop('Cabin',axis=1)
newdf.corr()
```

```python
nan_count_per_column = newdf.isnull().sum()
print("NaN count per column:")
print(nan_count_per_column)
```

```python
sklearn.linear_model.LassoLarsCV(cv=890).fit(newdf,Y).get_params()
```
